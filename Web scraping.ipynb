```python
import requests
import pandas as pd
```

```python
url = 'https://www.amazon.com/gp/bestsellers/?ref_=nav_cs_bestsellers'
#Amazon is likely blocking non-browser traffic. By adding the user agent string, you fool it into thinking that your script is a browser making a request.
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36',}
html_data = requests.get(url,headers=headers).text
```

```python
!pip install beautifulsoup4
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html_data,'html.parser')
```

```python
Item_a_tags = soup.find_all('a',{'class':'a-link-normal aok-block'})
```

```python
Item_name_tags = soup.find_all('div',{'class':'p13n-sc-truncate-desktop-type2  p13n-sc-truncated'})
```

```python
Item_price_tags = soup.find_all('span',{'class':'_cDEzb_p13n-sc-price_3mJ9Z'})
```

```python
Item_rating_tags = soup.find_all('i',{'class':'a-icon a-icon-star-small a-star-small-4-5 aok-align-top'})
```

```python
Item_links = []
for link in Item_a_tags:
    Item_links.append('https://www.amazon.com/'+link['href'])
```

```python
Item_prices = []
for price in Item_price_tags:
    Item_prices.append(price.text)
```

```python
Item_ratings = []
for rating in Item_rating_tags:
    Item_ratings.append(rating.text.split()[0])
```

```python
df = pd.DataFrame(list(zip(Item_links,Item_prices,Item_ratings)),columns=['Item Link','Price','Rating out of 5'])
```

```python
df.to_csv('Amazon bestseller.csv')
```
